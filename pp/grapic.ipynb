## Super Store deta anlaysis


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
file_path = r"C:\Users\HIMANSHU\OneDrive\Desktop\pp\superstore.csv"
superstore_data = pd.read_csv(file_path, encoding='latin1')
```

```python
# Q1: Most frequently used shipping mode
plt.figure(figsize=(8, 6))
sns.countplot(data=superstore_data, x='Ship Mode', hue='Ship Mode', palette='viridis', dodge=False, legend=False)
plt.title('Most Frequently Used Shipping Modes')
plt.xlabel('Ship Mode')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Q2: Total sales, profit, and quantity by region
region_summary = superstore_data.groupby('Region')[['Sales', 'Profit', 'Quantity']].sum().reset_index()
region_summary_melt = region_summary.melt(id_vars='Region', var_name='Metric', value_name='Value')
plt.figure(figsize=(10, 6))
sns.barplot(data=region_summary_melt, x='Region', y='Value', hue='Metric', palette='coolwarm')
plt.title('Sales, Profit, and Quantity by Region')
plt.tight_layout()
plt.show()
```

```python
# Q3: Customer with the highest total sales
customer_sales = superstore_data.groupby('Customer Name')['Sales'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
customer_sales.plot(kind='bar', color='teal')
plt.title('Top 10 Customers by Sales')
plt.xlabel('Customer Name')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Q4: Sales distribution across states
plt.figure(figsize=(14, 7))
sns.barplot(data=superstore_data, x='State', y='Sales', estimator=sum, hue='State', dodge=False, palette='viridis', legend=False)
plt.title('Sales Distribution Across States')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Q5: Profit by product categories
category_profit = superstore_data.groupby('Category')['Profit'].sum()
category_profit.plot(kind='bar', color='orange', figsize=(8, 6))
plt.title('Profit by Product Categories')
plt.xlabel('Category')
plt.ylabel('Profit')
plt.tight_layout()
plt.show()
```

```python
# Q6: Relationship between discount and profit
plt.figure(figsize=(8, 6))
sns.scatterplot(data=superstore_data, x='Discount', y='Profit', hue='Category', palette='viridis')
plt.title('Discount vs Profit')
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.tight_layout()
plt.show()
```

```python
# Q7: Sub-categories with the highest sales
sub_category_sales = superstore_data.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False).head(10)
sub_category_sales.plot(kind='bar', color='purple', figsize=(10, 6))
plt.title('Top 10 Sub-Categories by Sales')
plt.xlabel('Sub-Category')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()
```

```python
# Q8:  Total profit by shipping mode
ship_mode_profit = superstore_data.groupby('Ship Mode')['Profit'].sum()

# Plot total profit by shipping mode
ship_mode_profit.plot(kind='bar', color='blue', figsize=(8, 6))
plt.title('Total Profit by Shipping Mode')
plt.xlabel('Ship Mode')
plt.ylabel('Profit')
plt.tight_layout()
plt.show()
```

```python
# Q9: Cities contributing the most to sales
city_sales = superstore_data.groupby('City')['Sales'].sum().sort_values(ascending=False).head(10)
city_sales.plot(kind='bar', color='blue', figsize=(10, 6))
plt.title('Top 10 Cities by Sales')
plt.xlabel('City')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()
```

```python
# Q10: Most popular product by sales
product_sales = superstore_data.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
product_sales.plot(kind='bar', color='magenta', figsize=(10, 6))
plt.title('Top 10 Products by Sales')
plt.xlabel('Product Name')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()
```

```python
# Q11: Correlation between quantity and profit
plt.figure(figsize=(8, 6))
sns.scatterplot(data=superstore_data, x='Quantity', y='Profit', hue='Category', palette='cool')
plt.title('Quantity vs Profit')
plt.xlabel('Quantity')
plt.ylabel('Profit')
plt.tight_layout()
plt.show()
```

```python
# Q12: Profit by customer segment
segment_profit = superstore_data.groupby('Segment')['Profit'].sum()
segment_profit.plot(kind='bar', color='cyan', figsize=(8, 6))
plt.title('Profit by Customer Segment')
plt.xlabel('Segment')
plt.ylabel('Profit')
plt.tight_layout()
plt.show()
```

```python
# Q13: Average profit per order by region
region_avg_profit = superstore_data.groupby('Region')['Profit'].mean()
region_avg_profit.plot(kind='bar', color='lime', figsize=(8, 6))
plt.title('Average Profit per Order by Region')
plt.xlabel('Region')
plt.ylabel('Average Profit')
plt.tight_layout()
plt.show()
```

```python
# Q14: Sales distribution across postal codes
postal_sales = superstore_data.groupby('Postal Code')['Sales'].sum().sort_values(ascending=False).head(10)
postal_sales.plot(kind='bar', color='pink', figsize=(10, 6))
plt.title('Top 10 Postal Codes by Sales')
plt.xlabel('Postal Code')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()
```

```python
# Q15: Top 10 customers based on profit
customer_profit = superstore_data.groupby('Customer Name')['Profit'].sum().sort_values(ascending=False).head(10)
customer_profit.plot(kind='bar', color='coral', figsize=(10, 6))
plt.title('Top 10 Customers by Profit')
plt.xlabel('Customer Name')
plt.ylabel('Profit')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
# Q16: States with the highest discounts applied
state_discount = superstore_data.groupby('State')['Discount'].mean().sort_values(ascending=False).head(10)
state_discount.plot(kind='bar', color='brown', figsize=(10, 6))
plt.title('Top 10 States by Average Discount')
plt.xlabel('State')
plt.ylabel('Average Discount')
plt.tight_layout()
plt.show()
```

```python
# Q17: Profit-to-sales ratio for each product category
category_ratio = (superstore_data.groupby('Category')['Profit'].sum() / superstore_data.groupby('Category')['Sales'].sum())
category_ratio.plot(kind='bar', color='olive', figsize=(8, 6))
plt.title('Profit-to-Sales Ratio by Category')
plt.xlabel('Category')
plt.ylabel('Profit-to-Sales Ratio')
plt.tight_layout()
plt.show()
```

```python
# Top-performing regions by sales and profit
# Group by 'Region' and calculate total sales and profit
region_perf = superstore_data.groupby('Region')[['Sales', 'Profit']].sum()

# Plot side-by-side bar charts for sales and profit
region_perf.plot(kind='bar', figsize=(10, 6), color=['orange', 'purple'])
plt.title('Sales and Profit by Region')
plt.xlabel('Region')
plt.ylabel('Value')
plt.tight_layout()
plt.show()
```

```python
# Q20: Top-performing product IDs by profit
product_profit = superstore_data.groupby('Product ID')['Profit'].sum().sort_values(ascending=False).head(10)
product_profit.plot(kind='bar', color='steelblue', figsize=(10, 6))
plt.title('Top 10 Product IDs by Profit')
plt.xlabel('Product ID')
plt.ylabel('Profit')
plt.tight_layout()
plt.show()
```

```python
#Average discount by sub-category
# Group by 'Sub-Category' and calculate average discount
sub_category_discount = superstore_data.groupby('Sub-Category')['Discount'].mean().sort_values()

# Plot average discount by sub-category
sub_category_discount.plot(kind='bar', color='teal', figsize=(10, 6))
plt.title('Average Discount by Sub-Category')
plt.xlabel('Sub-Category')
plt.ylabel('Average Discount')
plt.tight_layout()
plt.show()
```

```python
#Q24: Top 10 profitable citie
# Group by 'City' and calculate total profit
city_profit = superstore_data.groupby('City')['Profit'].sum().sort_values(ascending=False).head(10)

# Plot top 10 profitable cities
city_profit.plot(kind='bar', color='cyan', figsize=(10, 6))
plt.title('Top 10 Profitable Cities')
plt.xlabel('City')
plt.ylabel('Profit')
plt.tight_layout()
plt.show()
```

```python
#Q26: Distribution of sales across segments
# Plot sales distribution across segments
plt.figure(figsize=(8, 6))
sns.boxplot(data=superstore_data, x='Segment', y='Sales', palette='coolwarm')
plt.title('Sales Distribution by Segment')
plt.xlabel('Segment')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()
```

```python
#Q27: Sub-category contribution to total profit
# Group by 'Sub-Category' and calculate total profit
sub_category_profit = superstore_data.groupby('Sub-Category')['Profit'].sum().sort_values()

# Plot sub-category contribution to profit
sub_category_profit.plot(kind='barh', color='orange', figsize=(10, 8))
plt.title('Profit Contribution by Sub-Category')
plt.xlabel('Profit')
plt.ylabel('Sub-Category')
plt.tight_layout()
plt.show()
```

```python
#Q28: Top-performing states by quantity sold
# Group by 'State' and calculate total quantity
state_quantity = superstore_data.groupby('State')['Quantity'].sum().sort_values(ascending=False).head(10)

# Plot top-performing states by quantity
state_quantity.plot(kind='bar', color='lime', figsize=(10, 6))
plt.title('Top 10 States by Quantity Sold')
plt.xlabel('State')
plt.ylabel('Quantity Sold')
plt.tight_layout()
plt.show()
```

```python
#Q29: Percentage contribution of each category to sales
# Calculate percentage contribution
category_sales_pct = (superstore_data.groupby('Category')['Sales'].sum() / superstore_data['Sales'].sum()) * 100

# Plot percentage contribution
category_sales_pct.plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'lightgreen', 'coral'], figsize=(8, 8))
plt.title('Category Contribution to Sales')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
#Q30: Regions with the highest average sales per order
# Calculate average sales per order by region
avg_sales_per_order = superstore_data.groupby('Region')['Sales'].mean()

# Plot average sales per order
avg_sales_per_order.plot(kind='bar', color='violet', figsize=(8, 6))
plt.title('Average Sales per Order by Region')
plt.xlabel('Region')
plt.ylabel('Average Sales')
plt.tight_layout()
plt.show()
```
